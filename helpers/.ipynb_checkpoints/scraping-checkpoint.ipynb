```python
ID = re.findall('[0-9]+',soup.find('a', {'rel':'nofollow', 'itemprop':'image'})['href'])[0]
ISBN = re.sub('\s+', '', tree.xpath('//*[@id="bookDataBox"]/div[2]/div[2]/text()')[0].replace("\n", ''))
title = re.sub('\s+', ' ',tree.xpath('//*[@id="bookTitle"]/text()')[0].replace("\n", ''))
author  = tree.xpath('//*[@id="bookAuthors"]/span[2]/div/a/span/text()')[0]
pulisher = re.sub('\s+', ' ', tree.xpath('//*[@id="details"]/div[2]/text()')[0].split('\n')[3])
date_pub = re.sub('\s+', ' ', tree.xpath('//*[@id="details"]/div[2]/text()')[0].split('\n')[2])
count_rating = re.findall('[0-9]+,*[0-9]+'," ".join(tree.xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[2]/a[2]/text()')).replace('\n',''))[0]
count_review = re.findall('[0-9]+,*[0-9]+'," ".join(tree.xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[2]/a[3]/text()')).replace('\n',''))[0]
description = soup.find('div', {'class':'readable stacked'}).text.replace('\n', ' ')
count_page = tree.xpath('//*[@id="details"]/div[1]/span[3]/text()')[0].split(' ')[0]
genres = []
for  i in range(10):
    g = dict()
    g['gendre'] = tree.xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div[5]/div[6]/div/div[2]/div')[0][i][0][0].text
    g['count_users'] = re.sub('\s+', " ",tree.xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div[5]/div[6]/div/div[2]/div')[0][i][1][0].text)
    genres.append(g)
book = info_Book(ID, ISBN, title, author, pulisher, date_pub, count_rating, count_review, description, count_page, genres)
```

```python
def info_Book(ID, ISBN, title, author, pulisher, date_pub, count_rating, count_review, description, count_page, Genres):
    dic_book = dict()
    dic_book['ID']          = ID
    dic_book['ISBN']        = ISBN
    dic_book['title']       = title
    dic_book['author']      = author
    dic_book['pulisher']    = pulisher
    dic_book['date_pub']    = date_pub
    dic_book['count_rating']= count_rating
    dic_book['count_review']= count_review
    dic_book['description'] = description.replace('(less)', '')
    dic_book['count_page']  = count_page
    dic_book['Genres']      = Genres
    return dic_book
```

```python
def rating_review(ID_book, ID_user, dic_review, info_user, Rating_book):
    dic_rating_reviewr = dict()
    dic_rating_reviewr['ID_book'] = ID_book
    dic_rating_reviewr['ID_user'] = ID_user
    dic_rating_reviewr['dic_review'] =  dic_review
    dic_rating_reviewr['info_user'] = info_user
    dic_rating_reviewr['Rating_book'] = Rating_book * 2
    return dic_rating_reviewr
def info_reviews(Date, Review, count_like_review, see_review):
    dic_review = dict()
    dic_review['Date']   = Date
    dic_review['Review'] = Review
    dic_review['count_like_review'] = count_like_review
    dic_review['see_review'] = see_review
    return dic_review
def Info_user(ID, Name, Link_profile):
    dic_user = dict()
    dic_user['ID'] = ID
    dic_user['Name']   = Name
    dic_user['Link_profile']   = Link_profile
    return dic_user
```
