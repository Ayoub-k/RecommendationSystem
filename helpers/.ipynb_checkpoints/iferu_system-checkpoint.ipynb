```python
# je vais faire une recommandation base sur le contenu on utilisant TfidfVectorizer et recommandation par chaine
```

```python
def recommandation_(title, chaine):
    global df_final
    df_final_ = df_final[df_final['nom_en'] == chaine]
    tfidf_matrix = tf.fit_transform(df_final_['courte_description'])
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
    df_final_ = df_final_.reset_index()
    titles = df_final_['titre']
    indices = pd.Series(df_final_.index, index=df_final_['titre'])
    index = indices[title]
    sim_scores = list(enumerate(cosine_sim[index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:60]
    livre_indices = [i[0] for i in sim_scores]
    print(livre_indices)
    return titles.iloc[livre_indices]
```
