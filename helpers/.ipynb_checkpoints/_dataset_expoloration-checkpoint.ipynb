```python
combiner_livres_rating = pd.merge(ratings, books, on='ISBN')
```

```python
combiner_livres_rating
```

```python
combiner_livres_rating.columns
```

```python
columns = ['Year-Of-Publication', 'Publisher', 'Book-Author', 'Image-URL-S', 'Image-URL-M',
       'Image-URL-L']
```

```python
combiner_livres_rating = combiner_livres_rating.drop(columns, axis=1)
```

```python
combiner_livres_rating
```

```python
combiner_livres_rating = combiner_livres_rating.dropna(axis = 0, subset = ['Book-Title'])
```

```python
combiner_livres_rating
```

```python
book_ratingCount = (combiner_livres_rating.
                    groupby(by = ['Book-Title'])['Book-Rating'].
                    count().
                    reset_index().
                    rename(columns = {'Book-Rating': 'totaleRatingCount'})
                    [['Book-Title', 'totaleRatingCount']]
                   )
```

```python
book_ratingCount
```

```python
rating_with_totalRatingCount = combiner_livres_rating.merge(book_ratingCount, left_on = 'Book-Title', right_on = 'Book-Title')
```

```python
rating_with_totalRatingCount
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
book_ratingCount['totaleRatingCount'].describe()
```

```python
# Le livre médian n'a été évalué qu'une seule fois. Regardons le haut de la distribution:
```

```python
book_ratingCount['totaleRatingCount'].quantile(np.arange(.9, 1, .01))
```

```python
pop_threshold = 50
rating_popular_book = rating_with_totalRatingCount.query('totaleRatingCount >= @pop_threshold ')
rating_popular_book
```

```python
rating_popular_book.drop_duplicates()
```

```python
# Filtrer sur les utilisateurs 
```

```python
combined = rating_popular_book.merge(users, left_on='User-ID', right_on='User-ID', how="left")
```

```python
combined
```

```python
combined = combined.drop_duplicates()
combined
```

```python
# Implémentation de kNN
```

```python
# Nous convertissons notre tableau en une matrice 2D et remplissons les valeurs manquantes
# par des zéros (puisque nous calculerons les distances entre les vecteurs de notation).
# Nous transformons ensuite les valeurs (notations) de la matrice de données en une matrice
# scipy clairsemée pour des calculs plus efficaces.
```

```python
# Trouver les voisins les plus proches
```

```python
combined = combined.drop('Age', axis=1)
```

```python
user_rating_pivot = combined.pivot_table(index='User-ID', columns='Book-Title', values='Book-Rating').fillna(0)
# combined.pivot_table(index='User-ID', columns='Book-Title').fillna(0)
user_rating_pivot 
```

```python
user_rating_matric = csr_matrix(user_rating_pivot.values)
```

```python
user_rating_matric
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(user_rating_matric)
```

```python
# Test our model and make some recommendations:
query_index = np.random.choice(user_rating_pivot.shape[0])
distance, index = knn_model.kneighbors(user_rating_pivot.iloc[query_index, :].values.reshape(1, -1), n_neighbors=7)
```

```python
query_index
```

```python
# user_rating_pivot.iloc[query_index, :].values.reshape(1, -1)
for i in range(len(distance.flatten())):
    if i==0:
        print('recommendation for {}'.format(user_rating_pivot.columns[query_index]))
    else:
        print('{} : {}, with distance of {}'.format(i, user_rating_pivot.columns[index.flatten()[i]], distance.flatten()[i] )) 
```

```python
user_rating_pivot.index[query_index]
```

```python
distance.flatten()
```

```python
user_rating_pivot.columns[index.flatten()]
```

```python
# Collaborative Filtering Using Matrix Factorization
```

```python
user_rating_pivot_2 = combined.pivot_table(index='User-ID', columns='Book-Title', values='Book-Rating').fillna(0)
```

```python
user_rating_pivot_2.head(1)
```

```python
 user_rating_pivot_2.shape
```

```python
X = user_rating_pivot_2.values.T
```

```python
X.shape
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)
```

```python
matric = SVD.fit_transform(X)
matric.shape
```

```python
import warnings
```

```python
warnings.filterwarnings('ignore', category=RuntimeWarning )
```

```python
corr = np.corrcoef(matric)
```
